```python
# import library
import pandas as pd
```

```python
# load the dataset
df = pd.read_csv('table.csv')
df
```

```python
# rename multiple columns
df = df.rename(columns={'Mat':'Matches_played',
                        'Inns':'Innings_batted',
                        'NO':'Not_outs','Runs':'Runs_scored',
                        'HS':'Highest_inns_score',
                        'Ave':'Avg',
                        'BF':'Balls_faced',
                        'SR':'Batting_strike_rate'
                       })
```

```python
# return the very first rows
df.head()
```

```python
# check for null values
df.isnull().any()
```

```python
# replace null values with zero
df['Balls_faced'] = df['Balls_faced'].fillna(0)
df['Batting_strike_rate'] = df['Batting_strike_rate'].fillna(0)
```

```python
# check if replaced
df[df['Player'] == 'CL Walcott (WI)']
```

```python
# check for duplicates
df.duplicated()
```

```python
# extract rows which has duplicate values
df[df['Player'].duplicated() == 1]
```

```python
# remove duplicates
df = df.drop_duplicates()
```

```python
# # check for duplicates again
df.duplicated().any()
```

```python
# split span column into start date and end date columns
df[['Start_date','End_date']] = df['Span'].str.split('-',expand=True)
df
```

```python
# drop span column
df = df.drop(['Span'], axis=1)
```

```python
df.head()
```

```python
# insert a country column using player column
df['Player'].str.split(pat='(')
```

```python
df['Country'] = df['Player'].str.split(pat='(').str[1]
```

```python
df['Country'] = df['Country'].str.split(pat=')').str[0]
```

```python
df['Country']
```

```python
df
```

```python
# modify data in player column
df['Player'] = df['Player'].str.split(pat='(').str[0]
```

```python
df.head()
```

```python
# remove unnecessary symbols from the data
df['Highest_inns_score'] = df['Highest_inns_score'].str.split(pat='*').str[0]
```

```python
df['Balls_faced'] = df['Balls_faced'].str.split(pat='+').str[0]
```

```python
# check for null values
df.isnull().any()
```

```python
# identify the rows with null values
df[df['Balls_faced'].isna()==1]
```

```python
# replacing null values with zero
df['Balls_faced'] = df['Balls_faced'].fillna(0)
```

```python
# display the data types of each column
df.dtypes
```

```python
# convert data types
df = df.astype({'Highest_inns_score':'int',
                'Balls_faced':'int',
                'Start_date':'int',
                'End_date':'int'
               })
```

```python
# add a career lenght column
df['Career_length'] = df['End_date'] - df['Start_date']
```

```python
# Q1- What is the average of career length?
df['Career_length'].mean()
```

```python
# Q2- What is the average of Batting_strike_rate for the players who played more than 10 years?
df[df['Career_length'] > 10]['Batting_strike_rate'].mean()
```

```python
# Q3- What is the number of players who played before 1990?
df[df['Start_date'] < 1990]['Player'].count()
```

```python
# What is the maximum value of Highest_inns_score by Country?
df.groupby('Country')['Highest_inns_score'].max().to_frame('Highest_inns_score').reset_index().sort_values('Highest_inns_score', ascending=False)
```
